```python
import pandas as pd
import torch
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

path = "datasets/"
values = torch.load(path +"values.pt")
datetimes = torch.load(path + "datetimes.pt")
print(values.shape)
```

```python
df = pd.DataFrame(values[:,0,:].numpy(), columns = datetimes)
df.head()
```

```python
plt.scatter(df.mean(axis=1), df.max(axis=1))
plt.xlabel("Mean")
plt.ylabel("Max")
plt.title("Individuals statistitics distributions")
```

```python
outliers = df[df.max(axis=1) > 1e5].reset_index(drop=True)
outliers.shape
```

```python
plt.plot(outliers.iloc[0])
```

```python
plt.plot(outliers.iloc[0][datetimes[-8000:-5000]])
```
